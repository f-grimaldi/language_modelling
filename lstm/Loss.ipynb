```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tr_loss, val_loss = [], []
with open('loss.txt', 'r') as file:
    for line in file: 
        tpos = line.find('A')
        vpos = line.find('V')
        if (tpos != -1) and (vpos != 1):
            try:
                tr_loss.append(float(line[tpos+14:tpos+24])) 
                val_loss.append(float(line[vpos+17:vpos+27]))
            except ValueError:
                continue
```

```python
plt.plot(np.log(val_loss), c='orange', label='Validation')
plt.plot(np.log(tr_loss), c='steelblue', label='Train')
plt.legend()
plt.title('MSE Loss')
plt.xlabel('Epochs')
plt.ylabel('Log loss')
plt.grid()
plt.show()
```
